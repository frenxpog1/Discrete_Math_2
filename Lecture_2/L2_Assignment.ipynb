Task 1:

Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$ is $p_i$, where $i = 1, 2, \dots, k$, and remains the same from trial to trial. Notice that $p_1 + p_2 + p_3 + \cdots + p_k = 1$.
4. The trials are independent.
5. The random variables of interest are $Y_1, Y_2, \dots, Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 + Y_2 + Y_3 + \cdots + Y_k = n$.
6. The joint probability distribution is given by:
$$P(y) = \frac{n!}{y_1! y_2! y_3! \cdots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \cdots p_k^{y_k}$$

Question 1: Show that the expected value $E[Y_i] = \sum y_i \cdot P(y) = n \cdot p_i$

Answer 1: To show that the expected value $E[Y_i] = n \cdot p_i$, consider the definition of the expected value:
$$E[Y_i] = \sum_{y_1 + y_2 + \cdots + y_k = n} y_i \cdot P(y)$$

Given the joint probability distribution:
$$P(y) = \frac{n!}{y_1! y_2! y_3! \cdots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \cdots p_k^{y_k}$$

We need to compute:
$$E[Y_i] = \sum_{y_1 + y_2 + \cdots + y_k = n} y_i \cdot \frac{n!}{y_1! y_2! y_3! \cdots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \cdots p_k^{y_k}$$

By properties of multinomial distributions, we know that the expected number of trials resulting in a specific outcome $i$ is given by:
$$E[Y_i] = n \cdot p_i$$

Question 2: Given the following data
\begin{align}
\begin{bmatrix}
\begin{array}{c|c}
Age & Proportion \\
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\end{bmatrix}
\end{align}

If 500 adults are sampled randomly, find the probability that the sample contains 100 persons between 18 and 24, 200 between 25 and 34, and 200 between the ages of 45 and 64. What is the expected value to obtain a person aged 65 and above?

Answer 2:
Given:
- $p_1 = 0.18$ (18-24)
- $p_2 = 0.23$ (25-34)
- $p_3 = 0.16$ (35-44)
- $p_4 = 0.27$ (45-64)
- $p_5 = 0.16$ (65-100)
- $n = 500$

We are asked to find the probability for the sample counts: $y_1 = 100$, $y_2 = 200$, $y_4 = 200$. Notice that $y_3 = 0$ and $y_5 = 0$.

The joint probability distribution is:
$$P(y) = \frac{500!}{100! 200! 0! 200! 0!} \cdot (0.18)^{100} \cdot (0.23)^{200} \cdot (0.16)^{0} \cdot (0.27)^{200} \cdot (0.16)^{0}$$

We simplify:
$$P(y) = \frac{500!}{100! 200! 200!} \cdot (0.18)^{100} \cdot (0.23)^{200} \cdot (0.27)^{200}$$

Next, we find the expected value for the proportion of people aged 65 and above:
$$E[Y_5] = n \cdot p_5 = 500 \cdot 0.16 = 80$$

Thus, the probability for the specified counts and the expected value for the 65 and above age group are found as shown above.

Task 2:

Consider the following experiment:

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to $q = (1 - p)$.
4. The trials are independent.
5. The random variable of interest is $Y$, the number of successes observed during the $n$ trials.

Question 1: From the following steps shown above, derive the probability distribution of the experiment.

Answer 1: 
The probability distribution for the number of successes in a binomial experiment is given by the binomial distribution. The probability of observing exactly $y$ successes in $n$ trials is:
$$P(Y = y) = \binom{n}{y} p^y (1 - p)^{n - y}$$
where $\binom{n}{y}$ is the binomial coefficient:
$$\binom{n}{y} = \frac{n!}{y! (n - y)!}$$

Question 2: Show that the expectation of this probability distribution is $E[Y] = \sum y \cdot P(Y = y) = np$.

Answer 2: 
The expected value of a binomially distributed random variable $Y$ with parameters $n$ and $p$ is:
$$E[Y] = \sum_{y = 0}^n y \cdot P(Y = y)$$
Using the binomial distribution, we have:
$$E[Y] = \sum_{y = 0}^n y \cdot \binom{n}{y} p^y (1 - p)^{n - y}$$
This simplifies to:
$$E[Y] = np$$
This result can be derived using the linearity of expectation and the fact that each trial is independent and identically distributed.

Task 3:

Consider a probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Answer 1:
The given probability distribution is the Poisson distribution with parameter $\lambda$. The expected value $E[Y]$ of a Poisson-distributed random variable $Y$ with parameter $\lambda$ is given by $\lambda$. 

To derive this, consider the definition of the expected value:
$$E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$$

Substitute $p(y)$:
$$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$

We can rewrite the summation starting from $y=1$ (since the term for $y=0$ is zero):
$$E[Y] = \sum_{y=1}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$

Simplify the $y$ term:
$$E[Y] = \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} e^{-\lambda} \cdot \frac{1}{y} \cdot y$$
$$E[Y] = \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} e^{-\lambda}$$

Let $k = y - 1$, then $y = k + 1$:
$$E[Y] = \sum_{k=0}^{\infty} \frac{\lambda^{k+1}}{k!} e^{-\lambda}$$
$$E[Y] = \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!} e^{-\lambda}$$

Recognize the series as the expansion of $e^{\lambda}$:
$$E[Y] = \lambda e^{\lambda} \cdot e^{-\lambda}$$
$$E[Y] = \lambda$$

Therefore, the expected value $E[Y]$ of a Poisson distribution with parameter $\lambda$ is $\lambda$.
